```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://www.bcci.tv/international/women/stats/t20i"
```

```python
headers = {
    "User-Agent": "Mozilla/5.0"
}
```

```python
response = requests.get(url, headers=headers)
response
```

```python
soup = BeautifulSoup(response.text, "html.parser")
```

```python
table = soup.find_all("table")[1]

rows = table.find_all("tr")

for i, row in enumerate(rows):
    cols = row.find_all("td")

    print(f"\nRow {i} | Total Columns: {len(cols)}")

    for j, col in enumerate(cols):
        h6 = col.find("h6")

        if h6:
            print(f"Column {j}: {h6.text.strip()}")
        else:
            print(f"Column {j}: No <h6> tag")
```

```python
table0 = soup.find_all("table")[0]
cols = table0.find_all("td")
```

```python
stats = {}

for td in cols:
    key = td.find("span").text.strip()
    value = td.find("p").text.strip()
    stats[key] = value
```

```python
matches = stats["Matches"]
innings = stats["Inns"]
average = stats["Avg."]
strike_rate = stats["SR"]
highest_score = stats["HS"]
fours = stats["4’s"]
sixes = stats["6’s"]
fifties = stats["50’s"]
hundreds = stats["100’s"]
runs = stats["Runs"]
```

```python
top_player = soup.find("div", class_="team-ranking-wrapper player")
rank = top_player.find("h1").text.strip().replace("#", "")
```

```python
name_div = top_player.find("div", class_="player-name-trw text-white mb-3")
player = " ".join(name_div.stripped_strings)
```

```python
data.insert(0, [
    rank,
    player,
    matches,
    innings,
    average,
    strike_rate,
    highest_score,
    fours,
    sixes,
    fifties,
    hundreds,
    runs
])

print(rank)
print(player)
print(matches)
print(innings)
print(average)
print(strike_rate)
print(highest_score)
print(fours)
print(sixes)
print(fifties)
print(hundreds)
print(runs)
```

```python
df = pd.DataFrame(data, columns=[
    "Rank",
    "Player",
    "Matches",
    "Innings",
    "Average",
    "Strike Rate",
    "Highest Score",
    "Fours",
    "Sixes",
    "Fifties",
    "Hundreds",
    "Runs"
])
```
