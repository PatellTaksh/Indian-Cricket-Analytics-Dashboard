```python
import pandas as pd
```

```python
men_test = pd.read_csv("../Data_Cleaning/Cleaned_CSV/Clean_Men_Test.csv")
men_odi = pd.read_csv("../Data_Cleaning/Cleaned_CSV/Clean_Men_ODI.csv")
men_t20i = pd.read_csv("../Data_Cleaning/Cleaned_CSV/Clean_Men_T20I.csv")

women_test = pd.read_csv("../Data_Cleaning/Cleaned_CSV/Clean_Women_Test.csv")
women_odi = pd.read_csv("../Data_Cleaning/Cleaned_CSV/Clean_Women_ODI.csv")
women_t20i = pd.read_csv("../Data_Cleaning/Cleaned_CSV/Clean_Women_T20I.csv")
```

```python
men_master = pd.merge(
    men_test,
    men_odi,
    on="Player",
    how="outer"
)

men_master = pd.merge(
    men_master,
    men_t20i,
    on="Player",
    how="outer"
)
men_master["Gender"] = "Male"
```

```python
women_master = pd.merge(
    women_test,
    women_odi,
    on="Player",
    how="outer"
)

women_master = pd.merge(
    women_master,
    women_t20i,
    on="Player",
    how="outer"
)
women_master["Gender"] = "Female"
```

```python
player_data = pd.concat(
    [men_master, women_master],
    ignore_index=True
)
```

```python
col = list(player_data.columns)

col[0], col[1] = col[1], col[0]
player_data = player_data[col]
```
